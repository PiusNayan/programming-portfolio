### Exercise 2 — Grade Calculator

```python
score = int(input("Enter your score here: "))
if score < 0 or score > 100:
    print("Invalid score.")
elif score >= 80:
    print("A")
elif score >= 70:
    print("B")
elif score >= 60:
    print("C")
elif score >= 50:
    print("D")
else:
    print("F")
```

## Exercise 4 — Number Guessing Game ⭐ (Mini Project)

Your first game!

Requirements:<br>
•Import <i>random</i> <br>
•Generate a secret number from 1–10<br>
•Ask the user to guess<br>
•Use a loop until they get it correct<br>
•Tell them if their guess is too high or too low<br>
•Display a congratulatory message when they win<br>

```python
import random
secret_number = random.randint(1,10)

guess = int(input("Guess any random number between 1-10. "))

atempts = 1

while guess != secret_number:
    if guess > secret_number:
        print(f'Incorrct, you guessed too high ')
    else:
        print(f'Incorrect, you guessed too low ')
    
    atempts += 1
    guess = int(input("Try again"))
print(f'Congratulations🎉, you guessed it in {atempts} attempts.')
```
